```python
from os.path import basename
import numpy as np
from roifile import  ImagejRoi
from roifile import ROI_TYPE
import random
import string
from zipfile import ZipFile
```

```python
def get_random_string(length,num_strings):
    # choose from all lowercase letter
    letters = string.ascii_lowercase
    op_str = []
    for i in range(num_strings):
        result_str = ''.join(random.choice(letters) for i in range(length))
        op_str.append(result_str)
    return op_str
    # print("Random string of length", length, "is:", result_str)

def GetRoiObject(
        points=None,
        roi_type = 5,
        name=None,
        position=None,
        index=None,
        c=None,
        z=None,
        t=None):
    roi_obj = ImagejRoi.frompoints(
        points=points,
        name=name,
        position=position,
        index=index,
        c=c,
        z=z,
        t=t)
    roi_obj.roitype = ROI_TYPE(roi_type)
    return roi_obj

def SaveAsRoi(
        points=None,
        roi_type =5,
        name=None,
        position=None,
        index=None,
        c=None,
        z=None,
        t=None,
        filename = None):
    
    
    if points is None:
        return None
    if filename is None:
        filename = get_random_string(10,1)
    roi = GetRoiObject(points,roi_type,
        name,
        position,
        index,
        c,
        z,
        t)
    out = roi.tobytes()
    roi.tofile(filename)
   
    
```

```python
def LoadDend(Dir):

    """
    Input:
            Dir (String)   : Super directory we are looking at
    Output:
            DendArr  (numpy array) : location of dendrite
        
    Function:
            Read npy files to obtain dendrites
    """
    print(Dir)
    DendArr = []
    DendArr_names = []
    soma = []
    for x in os.listdir(Dir):
        if('AllDendrites' in x):
            DendArr = np.load(Dir+x,allow_pickle=True)
        # elif('Dendrite' in x):
        #     DendArr.append(np.load(Dir+x))
        #     num = x.split("Dendrite")[1].split(".npy")[0]
        #     DendArr_names.append(num)
        elif("Soma" in x) and x.endswith(".npy"):
            soma.append(np.load(Dir+x))
        
    return DendArr,soma,DendArr_names
```

```python
def CreateFolderRecursive(folder):
    Path(folder).mkdir(parents=True, exist_ok=True)
```

```python
rootpath = "/Users/surbhitwagle/Desktop/Surbhit/Work/PhD/2020/PhD/MPIBR/PhD-Project/Experimental_collab/Anne-Sophie/Gria2/"
t = 0
for idx,file in enumerate(os.listdir(rootpath)):
    if file.startswith("cell"):
        t += 1
        cell_dir = rootpath+file+"/"
        DendArr,soma_arrs,DendArr_names = LoadDend(cell_dir)
        Roi_dir = cell_dir+"Rois/"
        CreateFolderRecursive(Roi_dir)
        file_list = []
        soma_file_list = []
        for dxd, dend in enumerate(DendArr):
            print(dend.shape)
            of_name = "Dendrite_{}.roi".format(dxd+1)
            SaveAsRoi(dend,name = "Dendrite_{}".format(dxd+1),filename=Roi_dir+of_name)
            file_list.append(of_name)
        for sxd,soma_arr in enumerate(soma_arrs):
            somaf_name = "Soma_{}.roi".format(sxd+1)
            SaveAsRoi(soma_arr,roi_type=0,name = "Soma_{}".format(sxd+1),filename=Roi_dir+somaf_name)
            soma_file_list.append(somaf_name)
        if len(file_list) > 1:
            with ZipFile(Roi_dir+'Dendrites.zip', mode='w') as zf:
                for f in file_list:
                    pf = Roi_dir+f
                    zf.write(pf,basename(pf))
                zf.close()
        if len(soma_file_list) > 1:
            with ZipFile(Roi_dir+'somas.zip', mode='w') as zf:
                for f in soma_file_list:
                    pf = Roi_dir+f
                    zf.write(pf,basename(pf))
                zf.close()
        RoiSet = file_list + soma_file_list
        if len(RoiSet) > 1:
            with ZipFile(Roi_dir+'RoiSet.zip', mode='w') as zf:
                for f in RoiSet:
                    pf = Roi_dir+f
                    zf.write(pf,basename(pf))
                zf.close()
```
